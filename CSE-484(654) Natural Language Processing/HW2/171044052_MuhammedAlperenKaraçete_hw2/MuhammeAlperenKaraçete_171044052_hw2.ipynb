```python
def build_ngram_model(text, n):
# Create a defaultdict to store the n-grams
    n_grams = defaultdict(int)

# Iterate over the list of words and create the 1-2-3 gram-s
    for i in range(len(text)-(n-1)):
        n_gram = tuple(text[i:i+n])
        n_grams[n_gram] += 1

    n_gram_counts = good_turing_smoothing(n_grams)    

    ngram_probs = {}
    total_count = sum(n_gram_counts.values())
    for ngram, count in n_gram_counts.items():
        ngram_probs[ngram] = count / total_count

    
    print_dict_table(n_grams)
    
    return ngram_probs
```

```python
def good_turing_smoothing(ngrams):
  # Count the frequency of each n-gram
    ngram_counts = Counter(ngrams)
    
  # Calculate the frequency of each count
    count_freqs = Counter(ngram_counts.values())
    
  # Calculate the smoothed probabilities
    smoothed_probs = {}
    for ngram, count in ngram_counts.items():
        smoothed_probs[ngram] = (count + 1) * count_freqs[count+1] / count_freqs[count]
    return smoothed_probs
```

```python
def print_dict_table(dictionary):
  # Find the maximum length of the keys and values
    max_key_len = max(len(str(key)) for key in dictionary)
    max_val_len = max(len(str(val)) for val in dictionary.values())
    
  # Print the table header
    print(f"{'Words':<{max_key_len}} {'Counts':<{max_val_len}}")
    print("-" * (max_key_len + max_val_len + 3))
    
  # Print the key-value pairs
    for key, val in dictionary.items():
        print(f"{str(key):<{max_key_len}} {str(val):<{max_val_len}}")
```

```python
from collections import defaultdict
from collections import Counter

with open("hw2.txt", "r",encoding="utf8") as f:
  # Read the contents of the file
    contents = f.read()

#Convert Turkish charachters to English
trans_table = str.maketrans("öüışğç", "ouisgc")
text = contents
text=text.lower() #Make lower all characters
text = text.translate(trans_table)
text = text.split()

# Set the number of words in the n-grams
n = 1 #This represents which gram we use, for 1-gram n=1, for 2-gram n=2 and for 3-gram n=3

build_ngram_model(text,n)
```
