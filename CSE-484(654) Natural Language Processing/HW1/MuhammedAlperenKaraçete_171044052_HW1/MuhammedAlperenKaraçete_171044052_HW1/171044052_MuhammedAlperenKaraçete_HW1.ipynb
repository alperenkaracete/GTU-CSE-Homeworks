```python
def smith_waterman_distance(line1, line2, match=3, substitution=-2, insertion=-2, deletion=-2):
    # switch sequences, so that line1 is the longer sequence to search for line2
    icoor=[]
    jcoor=[]
    if len(line2) > len(line1): line1, line2 = line2, line1
    # create the distance matrix
    mat = np.zeros((len(line2) + 1, len(line1) + 1))
    # iterate over the matrix column wise
    for i in range(1, mat.shape[0]):
        # iterate over the matrix row wise
        for j in range(1, mat.shape[1]):
            # set the current matrix element with the maximum of 4 different cases
            mat[i, j] = max(
                # negative values are not allowed
                0,
                # if previous character matches increase the score by match, else decrease it by substitution
                mat[i - 1, j - 1] + (match if line1[j - 1] == line2[i - 1] else substitution),
                # one character is missing in line2, so decrease the score by deletion
                mat[i - 1, j] + deletion,
                # one additional character is in line2, so decrease the scare by insertion
                mat[i, j - 1] + insertion
            )
            if (line1[j - 1]==line2[i - 1]):
                icoor.append(i-1)
                jcoor.append(j-1)
                
            
    # the maximum of mat is now the score, which is returned raw or normalized (with a range of 0-1)
    for k in range(len(icoor)):
        print("Matched lines are")
        print("1: ", line1[jcoor[k-1]],",j:Coordinate: ", jcoor[k-1])
        print("2: ", line2[icoor[k-1]],",i:Coordinate: ", icoor[k-1])
        print(mat[icoor[k]-5:icoor[k]+5,jcoor[k]-5:jcoor[k]+5])
        print("******************************************************")
    
```
